```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")


plt.figure(figsize=(8, 5))
sns.countplot(data=Pd, x="Category", order=Pd["Category"].value_counts().index, palette="viridis",legend="false")
plt.title("Product Distribution by Category", fontsize=14)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(Pd["Price"], bins=15, kde=True, color="skyblue")
plt.title("Product Price Distribution", fontsize=14)
plt.xlabel("Price (USD)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.boxplot(data=Pd, x="Category", y="Price", palette="Set2")
plt.title("Price Distribution by Category", fontsize=14)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Price (USD)", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python

Tr['TransactionDate'] = pd.to_datetime(Tr['TransactionDate'])


Tr['Month'] = Tr['TransactionDate'].dt.month_name()
Tr['DayOfWeek'] = Tr['TransactionDate'].dt.day_name()


summary = Tr.describe(include='all').transpose()
unique = Tr.nunique()


duplicates = Tr.duplicated().sum()


summary, unique, duplicates
```

```python

monthly_sales = Tr['Month'].value_counts()


daily_sales = Tr['DayOfWeek'].value_counts()


top_products_freq = Tr['ProductID'].value_counts().head(5)
top_products_revenue =Tr.groupby('ProductID')['TotalValue'].sum().nlargest(5)


top_customers_freq = Tr['CustomerID'].value_counts().head(5)
top_customers_spending = Tr.groupby('CustomerID')['TotalValue'].sum().nlargest(5)


avg_revenue_per_transaction = Tr['TotalValue'].mean()

monthly_sales, daily_sales, top_products_freq, top_products_revenue, top_customers_freq, top_customers_spending, avg_revenue_per_transaction
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_theme(style="whitegrid")


plt.figure(figsize=(10, 6))
monthly_sales.sort_index(ascending=True).plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Monthly Transactions", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Transactions", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()




```

```python
plt.figure(figsize=(10, 6))
daily_sales.sort_values(ascending=True).plot(kind='barh', color='lightgreen', edgecolor='black')
plt.title("Transactions by Day of the Week", fontsize=16)
plt.xlabel("Number of Transactions", fontsize=12)
plt.ylabel("Day of the Week", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
top_products_freq.sort_values(ascending=True).plot(kind='barh', ax=axes[0], color='coral', edgecolor='black')
axes[0].set_title("Top 5 Products by Frequency", fontsize=14)
axes[0].set_xlabel("Number of Transactions", fontsize=12)
axes[0].set_ylabel("Product ID", fontsize=12)
top_products_revenue.sort_values(ascending=True).plot(kind='barh', ax=axes[1], color='gold', edgecolor='black')
axes[1].set_title("Top 5 Products by Revenue", fontsize=14)
axes[1].set_xlabel("Revenue (₹)", fontsize=12)

plt.tight_layout()
plt.show()
```

```python





fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
top_customers_freq.sort_values(ascending=True).plot(kind='barh', ax=axes[0], color='lightblue', edgecolor='black')
axes[0].set_title("Top 5 Customers by Frequency", fontsize=14)
axes[0].set_xlabel("Number of Transactions", fontsize=12)
axes[0].set_ylabel("Customer ID", fontsize=12)

top_customers_spending.sort_values(ascending=True).plot(kind='barh', ax=axes[1], color='limegreen', edgecolor='black')
axes[1].set_title("Top 5 Customers by Spending", fontsize=14)
axes[1].set_xlabel("Total Spending (₹)", fontsize=12)

plt.tight_layout()
plt.show()
```
